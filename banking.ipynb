```python
cols=["member_id","annual_income","dti","installment","int_rate","loan_amount","total_acc","total_payment"]
```

```python
for i in cols:
    sns.displot(dfbank[i])
    plt.show()
```

```python
for i in cols:
    skew=dfbank[i].skew().round()
    print(skew,"-->",i)
```

```python
for i in cols:
    skew=dfbank[i].skew().round()
    if skew !=0:
        q1=dfbank[i].quantile(0.25)
        q3=dfbank[i].quantile(0.75)
        iqr=q3-q1
        up=q3+1.5*iqr
        lp=q3-1.5*iqr
        dfbank[i]=dfbank[i].map(lambda x : up if x>up else lp if x<lp else x)
    else :
        m=dfbank[i].mean()
        s=dfbank[i].std()
        ul=m+3*s
        ll=m-3*s
        dfbank[i]=dfbank[i].map(lambda x : ul if x>ul else ll if x<ll else x)
      
```

```python
sns.boxplot(dfbank)
plt.xticks(rotation=45)
plt.show()
```
